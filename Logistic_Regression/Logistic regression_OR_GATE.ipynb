# CODE FOR LOGISTIC REGRESSION 

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import sys
import math
```

```python
#dataset for OR GATE
x_init=[[0,0],[0,1],[1,0],[1,1]]
y=[0,1,1,1]
x_init=np.reshape(x_init,(4,2))
y=np.reshape(y,(4,1))
#appending comlumn for x^0 ie for bias weight
x=[]
for i in range(0,np.shape(x_init)[0]):
    x.append(float(1))
x=np.reshape(x,(np.shape(x_init)[0],1))    
x=np.append(x,x_init, axis=1)
y,x
```

```python
def sigmoid(z):
    return 1/(1+np.e**(-z))
```

```python
features=3
w=[]
for i in range(0,features):
    w.append(float(random.uniform(0,1)))
w=np.reshape(w,(features,1))
m=np.shape(x)[0]
```

```python
#deciding initial values for error(err) and learning rate(alpha)
err=1
alpha=0.1
iteration=0
J=[]
yh=np.zeros((m,1))
DJ=np.zeros((features,1))
DW=np.zeros((features,1))
```

```python
def cost(y,yh):
    return -np.mean(y*np.log(yh)+(1-y)*np.log(1-yh))   
```

```python
m=len(y)
i=0
while err>1.e-5:
    z=np.matmul(x,w)
    yh=sigmoid(z)
    J.append(cost(y,yh))
    dz=yh-y
    DW=1/m*np.matmul(x.T,dz)
    w=w-alpha*DW
    if(i>0):
        err=abs(J[i]-J[i-1])
    i=i+1    
        
```
